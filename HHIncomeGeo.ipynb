```python
import pandas as pd
import numpy as np
import folium as f
from folium.plugins import MarkerCluster
```

```python
hhgeo = pd.read_csv('household_income_kaggle.csv', encoding='ISO-8859-1')
hhgeo.head()
```

```python
#General Explore
hhgeo.shape
hhgeo.dtypes
```

```python
#Get Sample for Mapping

hhgeo = hhgeo.iloc[1:1000,:]
hhgeo.shape
```

```python
# From http://nbviewer.jupyter.org/github/python-visualization/folium/blob/master/examples/1000_MarkerCluster.ipynb

lon = hhgeo.Lon
lat = hhgeo.Lat

coordinates = list(zip(lon,lat))
popups = [format(i) for i in hhgeo.Place]

```

```python


m = f.Map(location = [np.mean(lat), np.mean(lon)], zoom_start=1)
m = m.add_child(MarkerCluster(locations=coordinates, popups=popups))
m
```
